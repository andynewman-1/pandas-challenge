```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
total_schools = school_data_complete["school_name"].nunique()
total_schools
```

```python
total_students = school_data_complete["student_name"].count()
total_students
```

```python
total_budget = school_data["budget"].sum()
total_budget
```

```python
av_maths_score = school_data_complete["maths_score"].mean()
av_maths_score
```

```python
av_reading_score = school_data_complete["reading_score"].mean()
av_reading_score
```

```python
pass_math = school_data_complete[school_data_complete["maths_score"] >= 50]["Student ID"].count()
pass_math
percent_pass_math = pass_math/total_students*100
percent_pass_math
```

```python
pass_read = school_data_complete[school_data_complete["reading_score"] >= 50]["Student ID"].count()
pass_read
percent_pass_read = pass_read/total_students*100
percent_pass_read
```

```python
pass_overall = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]["Student ID"].count()
pass_overall
percent_pass_overall = pass_overall/total_students*100
percent_pass_overall
```

# Local Government Area Summary

```python
#using dictionary of lists
lga_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                             "Total Students": [total_students],
                             "Total Budget": [total_budget],
                             "Average Maths Score": [av_maths_score],
                             "Average Reading Score": [av_reading_score],
                             "% Passing Maths": [percent_pass_math],
                             "% Passing Reading": [percent_pass_read],
                             "% Overall Passing": [percent_pass_overall]})
lga_summary_df["Total Budget"] = lga_summary_df["Total Budget"].map("${:,.2f}".format)
lga_summary_df
```

```python
#group the merged tables by school
groupedby_school = school_data_complete.groupby(["school_name"])
#groupedby_school.head()
```

```python
#define school type. Use of first stops it repeating itself
school_type = groupedby_school["type"].first()
#school_type
```

```python
#define total students by school
students_per_school = groupedby_school["Student ID"].count()
#students_per_school
```

```python
#define total budget by school
budget_per_school = groupedby_school["budget"].first()
```

```python
#define per student budget
student_budget = budget_per_school/students_per_school
```

```python
#define average maths score
avmath_score = groupedby_school["maths_score"].mean()
```

```python
#define average reading score
avread_score = groupedby_school["reading_score"].mean()
```

```python
#Calculate percent passing maths
perschool_pass_math = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["school_name"])["Student ID"].count()
#perschool_pass_math
percent_perschool_pass_math = perschool_pass_math/students_per_school*100
#percent_perschool_pass_math
```

```python
#Calculate percent passing maths
perschool_pass_read = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["school_name"])["Student ID"].count()
percent_perschool_pass_read = perschool_pass_read/students_per_school*100
```

```python
#calculate percent passing both
perschool_pass_overall = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)].groupby(["school_name"])["Student ID"].count()
#perschool_pass_overall
percent_perschool_pass_overall = perschool_pass_overall/students_per_school*100
#percent_perschool_pass_overall
```

 # School Summary

```python
#create summary df
school_summary_df = pd.DataFrame({"School Type": school_type,
                                  "Total Students": students_per_school,
                                 "Total School Budget": budget_per_school,
                                 "Per Student Budget": student_budget,
                                 "Average Maths Score": avmath_score,
                                 "Average Reading Score": avread_score,
                                 "% Passing Maths": percent_perschool_pass_math,
                                 "% Passing Reading": percent_perschool_pass_read,
                                 "% Overall Passing": percent_perschool_pass_overall})
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df
```

# Reading Score by Year

```python
#create data frame
read_year = pd.DataFrame({"Year 9": nine_read,
                         "Year 10": ten_read,
                         "Year 11": eleven_read,
                         "Year 12": twelve_read})
read_year
```

```python
#create school spending bins
bins = [0, 585, 630, 645, 680]
bins_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

# Scores by School Spending

```python

school_data_complete['Spending Ranges (Per Student)'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = bins_names)
by_spending = school_data_complete.groupby("Spending Ranges (Per Student)")

avg_math = by_spending["maths_score"].mean()

avg_read = by_spending["reading_score"].mean()
pass_math = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('Spending Ranges (Per Student)')['Student ID'].count()/by_spending['Student ID'].count()*100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('Spending Ranges (Per Student)')['Student ID'].count()/by_spending['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score'] >= 50)].groupby('Spending Ranges (Per Student)')['Student ID'].count()/by_spending['Student ID'].count()*100
    
# df build            
scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "% Overall Passing": overall})
    
scores_by_spend
```

# Scores by School Size

```python
#results by size
#create school size bins
size_bins = [0, 1000, 2000, 5000]
size_bins_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]


school_data_complete['School Size'] = pd.cut(school_data_complete['size'], size_bins, labels = size_bins_names)
by_size = school_data_complete.groupby("School Size")

avg_math = by_size["maths_score"].mean()

avg_read = by_size["reading_score"].mean()
pass_math = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('School Size')['Student ID'].count()/by_size['Student ID'].count()*100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('School Size')['Student ID'].count()/by_size['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score'] >= 50)].groupby('School Size')['Student ID'].count()/by_size['Student ID'].count()*100
    
# df build            
scores_by_size = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "% Overall Passing": overall})
    
scores_by_size
```
